### Questão 2
**Países com maior índice de rigidez nas restrições tem uma taxa de mortalidade menor?**

JUSTIFICATIVA:Verficar se as medidas restritivas auxiliou na diminuição da taxa de mortalidade

```python

indice_de_restricoes = dataset_covid.groupby(['location'])[['stringency_index']].median()
mortes_totais = dataset_covid.groupby(['location'])[['total_deaths']].sum()
populacao_por_location = dataset_covid.groupby(['location'])[['population']].median()

relacao_mortes_com_restricoes = pd.merge(indice_de_restricoes, mortes_totais, on='location').merge(populacao_por_location, on='location')

relacao_mortes_com_restricoes_ordenado=relacao_mortes_com_restricoes.sort_values(by=['stringency_index'], ascending=False).reset_index()
relacao_mortes_com_restricoes_ordenado[['location', 'stringency_index', 'total_deaths']].head(10)
```

###Questão 23

**Qual o top10 países com maior índice de ridigez do pais e qual foi seu nível de infecção por densidade populacional?**

```python
#Top 10 países com maior índice de rigidez
indice_rigidez=dataset_covid.groupby(['location'])[['stringency_index']].median().reset_index()
indice_rigidez.head(10)
```

```python
#Nível de Infecção por Densidade populacional

nivel_densidade_pop = dataset_covid.groupby(['location'])[['population_density']].median()
casos_totais = dataset_covid.groupby(['location'])[['total_cases']].max()

merged_densidade_casos = pd.merge(nivel_densidade_pop, casos_totais, on='location')
merged_densidade_casos['infecction_level'] = round(
    (merged_densidade_casos['total_cases'] / merged_densidade_casos['population_density']), 2)

```

```python
nivel_infeccao = pd.merge(merged_densidade_casos, indice_rigidez, on='location')

nivel_infeccao_ordenado = nivel_infeccao.sort_values(by=['stringency_index'], ascending=False).reset_index()
nivel_infeccao_ordenado.head(10)
```

**Conclusão:** considerando o nivel de infecção por densidade populacional (baseada no numeros de caso e a densidade populacional) dos 10 países que estabeleceram medidas mais rigidas de restrição,observa-se ainda assim a variação do nível de infecção.

###Questão 25

**Qual foi a quantidade de mortes por milhão de habitantes nos países sulamericanos?**

JUSTIFICATIVA:Provavelmente, ao comparar a quantidade de mortes por milhão de habitantes ajudará a ter um indicativo de quais políticas públicas seriam mais eficazes.

```python
sulamericano_covid=covid_death_df.loc[covid_death_df['continent']== 'South America'].groupby(['location'])[['total_deaths_per_million']].sum(numeric_only=True).reset_index()
sulamericano_covid
```

```python
sns.set_theme(style="whitegrid")

# Inicializando matplotlib figure
f, ax = plt.subplots(figsize=(6, 15))

sns.set_color_codes("pastel")
sns.barplot(x="total_deaths_per_million",
            y="location",
            data=sulamericano_covid.sort_values(by=['total_deaths_per_million'], ascending=False),
            label="total_deaths_per_million",
            color="b")
```

**Conclusão:** Através do gráfico observa-se que o Peru foi o país que teve o maior índice de mortes por milhão de habitante na América do Sul,acompanhado em seguida do Brasil e Argentina
